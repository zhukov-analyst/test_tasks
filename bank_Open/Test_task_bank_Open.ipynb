**ЗАДАНИЕ 1**  
Предложите 3 метрики первого уровня для мобильного и интернет-банка. Почему именно эти? 

**РЕШЕНИЕ**  
  
Модель метрик зависит от типа продукта, этапа его развития и целей продуктовой аналитики.  
Рассмотрим три группы самых общих метрик универсальных для мобильного и интернет-банка: 

**1. Цифровизация клиентов**
- доля цифровых клиентов: доля пользователей мобильного/интернет банка среди всех активных клиентов банка.
- конверсия новых активных клиентов в пользователей: доля новых активных клиентов, сразу ставшими пользователями мобильного/интернет банка.

Данные метрики позволят оценить охват среди клиентов банка и эффективность действий по его увеличению - необходимо стремится к 100%.

     
**2. Вовлеченность**
- Retention rate: коэффициент удержания - доля пользователей, которые продолжает использовать продукт через определённый период времени. 
- Churn rate: коэффициент оттока (метрика, обратная к retention rate).
- Daily Active Users / Monthly Active User: количество уникальных пользователей, которые используют мобильный/интернет банк.
- Session duration: средняя длительность сессии пользователей в мобильном/интернет банке.

Классические метрики оценки активности пользователей в разлиличных разрезах - эти метрики отражают общее "здоровье" канала.  


**3. Транзакции и продажи**
- доля транзакционных пользователей: доля пользователей, совершающие транзакции.
- частота транзакций: среднее количество транзакций на пользователя.
- сумма транзакций: средняя суммма транзакций на пользователя.
- конверсия в продажу: доля пользователей, увидивших информацию о дебетовом/кредитном продукте и совершивших покупку.

Этот набор метрик оценивает экономическую сторону эффективности канала. Здесь требуется погружение в специфику поведения отдельных аудиторных сегментов банка с целью максимизации их потребительской активности.

**ЗАДАНИЕ 2**  
Добавить индикатор, который будет выделять следующие значения:  
- Если otp для категории платежей (payment), то - 1  
- Если otp для категории переводов (transfer), но не для переводов с использованием счетов (acc), то - 2  
Все остальные заявки не должны попасть в результат выполнения запроса.  

```python
sql = '''
SELECT
otp.*
FROM (
        SELECT
        *,
        CASE
            WHEN category LIKE '%payment%' THEN '-1'
            WHEN category LIKE '%transfer%' AND category NOT LIKE '%acc%' THEN '-2'
            ELSE NULL
            END as flag
        FROM otp_session_id
    ) as otp
WHERE flag IN ('-1', '-2')

'''
```

```python
pd.read_sql(sql, con)
```

**Вывод**: промаркированы соответствующие категории, их в базе - 1588.

**ЗАДАНИЕ 3**  
Посчитать метрику Month-of-Month (прирост текущего месяца к предыдущему) по уникальным клиентам с кодами otp-login.

```python
sql = '''
with uniq_client_bymonth as (
    SELECT
    otp.month,
    count(DISTINCT otp.client_id) cnt_uniq
    FROM (
            SELECT
            *,
            cast(date_trunc('month',datetime) as date) as month
            FROM otp_session_id
            WHERE category = 'otp-login'
        ) as otp
    GROUP BY otp.month
)

SELECT
uc1.month as month_now, uc2.month as month_last,
(uc1.cnt_uniq * 1.0  / uc2.cnt_uniq * 1.0) as Month_of_Month 
FROM uniq_client_bymonth as uc1
CROSS JOIN uniq_client_bymonth as uc2
WHERE uc1.month != uc2.month AND uc1.month > uc2.month
'''
```

```python
pd.read_sql(sql, con)
```

**Вывод**: в июне уникальных клиентов было меньше, чем в мае (на ~30%).  
Запрос получился не самым оптимальным, на других данные он может ошибаться.  Есть где оптимизировать.

Одним запросом сформируйте:  
- Количество успешно введённых ОТП кодов в разрезе категории кода ОТП  
- Долю каждой категории по убыванию  
- Количество с накопительным итогом  
- Общее количество введённых ОТП кодов  
- Последний отчётный месяц (полный месяц, от текущего)  

```python
sql = '''
SELECT
f.category,
f.correct_cnt_cat,
f.share_cat,
sum(f.all_cnt) OVER (ORDER BY f.share_cat desc ROWS BETWEEN unbounded preceding and current row) as cum_cnt_cat,
f.all_cnt,
CASE 
    WHEN cast(date_trunc('month',f.date) as date) < cast(date_trunc('month', current_date) as date)
    THEN cast(date_trunc('month',f.date) as date)
    ELSE cast(date_trunc('month',f.date) as date) - interval '1 month'
    END as last_full_month
FROM (
        SELECT
        otp.category,
        sum(CASE WHEN status = 'CORRECT' THEN 1 ELSE 0 END) as correct_cnt_cat,
        count(otp.category)*1.0 / (SELECT count(*) FROM otp_session_id) as share_cat,
        count(otp.category) as all_cnt,
        max(otp.datetime) as date
        FROM otp_session_id as otp
        GROUP BY otp.category
    ) as f
ORDER BY f.share_cat desc
'''
```

```python
pd.read_sql(sql, con)
```

**Вывод**: получилось одним запросом вывести всю информацию.  
Единственное слабое место - последний отчетный месяц. Если данные не непрерывные, то могут быть ошибки (будет выведен месяц, которого на самом деле нет в данных).
