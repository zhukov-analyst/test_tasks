**Задача 1**  
Напишите запрос, который выведет пилотов, которые в качестве второго пилота в августе этого года ездили в аэропорт Шереметьево.

```python
sql = '''
SELECT 
DISTINCT f.second_pilot_id,
p.name
FROM flight as f
LEFT JOIN pilot as p ON f.second_pilot_id = p.pilot_id
WHERE (flight_dt BETWEEN '2021-08-01' AND '2021-08-31') 
    AND destination = 'Sheremetyevo International Airport'
    
'''

pd.read_sql(sql, con)
```

**Задача 2**  
Выведите пилотов старше 45 лет, которые совершили больше 30 полетов на грузовых самолетах.

Из-за специфики тестовых данных я в условиях вывода уменьшу кол-во полётов с 30 до 6 (максимальное значение в тестовых данных).

```python
sql = '''
SELECT 
ff.pilot_id,
ff.name
FROM (
        SELECT *
        FROM flight as f
        LEFT JOIN plane as pl ON pl.plane_id = f.plane_id
        LEFT JOIN pilot as p ON f.first_pilot_id = p.pilot_id
        WHERE cargo_flg = '1' AND p.age >= 45
        
        UNION ALL
        
        SELECT *
        FROM flight as f
        LEFT JOIN plane as pl ON pl.plane_id = f.plane_id
        LEFT JOIN pilot as p ON f.second_pilot_id = p.pilot_id
        WHERE cargo_flg = '1' AND p.age >= 45
    ) as ff
GROUP BY ff.pilot_id, ff.name
HAVING count(ff.cargo_flg) >= 6 
'''

pd.read_sql(sql, con)
```

**Задача 3**  
Выведите ТОП 10 пилотов-капитанов (first_pilot_id), которые перевезли наибольшее число пассажиров в этом году.

```python
sql = '''
SELECT
p.pilot_id,
p.name,
sum(pl.capacity)
FROM flight as f
LEFT JOIN plane as pl ON pl.plane_id = f.plane_id
LEFT JOIN pilot as p ON f.first_pilot_id = p.pilot_id
WHERE cargo_flg = '0'
GROUP BY p.pilot_id, p.name
ORDER BY sum(pl.capacity) desc
LIMIT 10
'''

pd.read_sql(sql, con)
```
